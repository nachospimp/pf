```python
import pandas as pd
import numpy as np
import yfinance as yf
import pandas_ta as ta
import plotly.graph_objects as go

# Descargar datos de Yahoo Finance para AMC
ticker = yf.Ticker('NVDA')
data = ticker.history(interval="5m", period="5d")

# Calcular indicadores: VWAP, RSI, Bandas de Bollinger
data["VWAP"] = ta.vwap(data.High, data.Low, data.Close, data.Volume)
data["RSI"] = ta.rsi(data.Close, length=16)
my_bbands = ta.bbands(data.Close, length=14, std=2.0)
data = data.join(my_bbands)

# Definir señales VWAPSignal y TotalSignal
VWAPsignal = [0] * len(data)
backcandles = 15

for row in range(backcandles, len(data)):
    upt = 1
    dnt = 1
    for i in range(row - backcandles, row + 1):
        if max(data.Open[i], data.Close[i]) >= data.VWAP[i]:
            dnt = 0
        if min(data.Open[i], data.Close[i]) <= data.VWAP[i]:
            upt = 0
    if upt == 1 and dnt == 1:
        VWAPsignal[row] = 3
    elif upt == 1:
        VWAPsignal[row] = 2
    elif dnt == 1:
        VWAPsignal[row] = 1

data['VWAPSignal'] = VWAPsignal

# Definir TotalSignal
def TotalSignal(row):
    if (data.VWAPSignal[row] == 2
        and data.Close[row] <= data['BBL_14_2.0'][row]
        and data.RSI[row] < 45):
        return 2

    if (data.VWAPSignal[row] == 1
        and data.Close[row] >= data['BBU_14_2.0'][row]
        and data.RSI[row] > 55):
        return 1

    return 0

TotSignal = [0] * len(data)
for row in range(backcandles, len(data)):
    TotSignal[row] = TotalSignal(row)

data['TotalSignal'] = TotSignal

# Calcular medias móviles simples (SMA)
data.ta.sma(length=20, append=True)
data.ta.sma(length=50, append=True)

# Crear una columna para los puntos de señal
data['pointposbreak'] = np.where(data['TotalSignal'] != 0, data['High'] + 1e-4, np.nan)

# Seleccionar datos para graficar
st = 0
dfp1 = data.iloc[st:st+350]

# Filtrar los datos donde el volumen es mayor que cero
dfp1 = dfp1[dfp1['Volume'] > 0]

# Asegurar que el índice sea de tipo datetime y ordenar por fecha
dfp1.index = pd.to_datetime(dfp1.index)
dfp1.sort_index(inplace=True)

# Imprimir los valores del RSI para los primeros 10 registros
print("Valores del RSI:")
print(dfp1['RSI'].tail(10))  # Muestra los primeros 10 valores del RSI

# Crear la figura en Plotly
fig = go.Figure(data=[
    go.Candlestick(x=dfp1.index,
                   open=dfp1['Open'],
                   high=dfp1['High'],
                   low=dfp1['Low'],
                   close=dfp1['Close'],
                   increasing_fillcolor='black',  # Relleno blanco para velas alcistas
                   increasing_line_color='green',  # Borde verde para velas alcistas
                   name="Candlestick"),
                   go.Scatter(x=dfp1.index, y=dfp1['VWAP'],
                              mode='lines',
                              line=dict(color='blue', width=1),
                              name="VWAP"),
                   go.Scatter(x=dfp1.index, y=dfp1['BBL_14_2.0'],
                              mode='lines',
                              line=dict(color='pink', width=1),
                              name="BBL"),
                   go.Scatter(x=dfp1.index, y=dfp1['BBM_14_2.0'],
                              mode='lines',
                              line=dict(color='lightblue', width=1),
                              name="BBM"),
                   go.Scatter(x=dfp1.index, y=dfp1['BBU_14_2.0'],
                              mode='lines',
                              line=dict(color='orange', width=1),
                              name="BBU"),
                   go.Scatter(x=dfp1.index, y=dfp1['SMA_20'],
                              mode='lines',
                              line=dict(color='cyan', width=1),
                              name="SMA 20"),
                   go.Scatter(x=dfp1.index, y=dfp1['SMA_50'],
                              mode='lines',
                              line=dict(color='purple', width=1),
                              name="SMA 50"),
                   go.Scatter(x=dfp1.index, y=dfp1['pointposbreak'],
                                    mode="markers",
                                    marker=dict(size=5, color="MediumPurple"),
                                    name="Signal")
])

# Configurar diseño y mostrar la figura
fig.update_layout(
    title='Gráfico de Velas con Indicadores y Señales',
    xaxis_title='Fecha',
    yaxis_title='Precio',
    height=1100,
    width=1200,
    template='plotly_dark'  # Fondo oscuro del gráfico
)

# Mostrar la figura
fig.show()
```

```python
import numpy as np
def pointposbreak(x):
    if x['TotalSignal']==1:
      return x ['High']+1e-4
    elif x['TotalSignal']==2:
      return x ['High']-1e-4
    else:
      return np.nan
df['pointposbreak'] = df.apply(lambda row: pointposbreak, axis=1)
```
